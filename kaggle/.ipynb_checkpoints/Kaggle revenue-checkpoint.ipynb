```python
def toX_y(aa):
    aa['days'] = pd.to_datetime(aa["Open Date"]) - pd.datetime.today()
    aa['days'] = pd.Series([x.days for x in aa['days']])


    b = aa
    b["other"] = b["City Group"] == "Other"
    b["IL"] = b["Type"] == "IL"

    b["otherIL"] = b.IL & b.other

    b["aft4000"] = b.days > -4000
    b["aft1500"] = b.days > -1500


    b["istanbul"] = b.City == "İstanbul"

    b["ankara"] = b.City == "Ankara"

    b["antalya"] = b.City == "Antalya"

    b["adana"] = b.City == "Adana"
    b["izmir"] = b.City == "İzmir"

    b["bursa"] = b.City == "Bursa"

    b["sakarya"] = b.City =="Sakarya"
    b["samsun"] = b.City =="Samsun"
    b["tekirdag"] = b.City =="Tekirdağ"

    places = "istanbul ankara antalya adana izmir bursa sakarya samsun tekirdag"
    #X = b[["P%s" % x for x in range(1,38)] + "other IL aft4000 aft1500 istanbul ankara".split()]
    cols = places.split() + "days other IL otherIL aft4000 aft1500 P28 P25 P17".split()
    #cols = cols[:-3] + ["P%s" % x for x in range(1, 38)]

    #bb = b[(b.revenue > p5) & (b.revenue < p95)]
    X = b[cols]
    #y = bb.revenue
    try:
        y = b.revenue
        p5 = np.percentile(y, 5)
        p95 = np.percentile(y, 95)
        return X,y
    except:pass

    return X

X,y = toX_y(aa)


#pd.DataFrame([b.groupby("City").count().P1, b.groupby("City").mean().revenue, b.groupby("City").std().revenue]).T


#b.describe()
```

```python
def plot(a, shift=0, length = 9):

    segments = [("All",a)]
    for group in ['Other', 'Big Cities']:
        for TYPE in "FC IL DT MB".split(): # FC: Food Court, IL: Inline, DT: Drive Thru, MB: Mobile
            segments.append((group + " " + TYPE, a[(a['City Group'] == group) & (a["Type"] == TYPE)]))

    for seg, data in segments:
        plt.figure(figsize=(20,2))
        for i, col in enumerate(["days"] + ['P%i' % (shift+x) for x in range(1,1 + length)]):
            plt.subplot(1,10, i+1)
            plt.gca()#.scatter(data[col], data.revenue)
            if len(data[col])>0: #sns.residplot(data[col]+np.random.randn(len(data[col]))/10**10, data.revenue, color="indianred", lowess=True);
                plt.gca().hist2d(data[col], data.revenue)
            plt.gca().get_xaxis().set_ticklabels([])
            plt.gca().get_yaxis().set_ticklabels([])
#            plt.axis('equal')
#            plt.gca().set_aspect('equal')
            plt.gca().set_xlabel('corr=%.2f ^' % np.corrcoef(data[col], data.revenue)[0,1])
            if i == 0:
                plt.gca().set_ylabel(seg)
            if seg == "All":
                plt.gca().set_title(col)
```
